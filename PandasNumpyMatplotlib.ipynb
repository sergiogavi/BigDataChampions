```python
class MiembroClub(object):
    """Clase que representa una Persona del club"""

    def __init__(self, dni, nombre, apellido, sexo):
        """Constructor de clase Persona"""
        self.dni = dni
        self.nombre = nombre
        self.apellido = apellido
        self.sexo = sexo

    def __str__(self):
        """Devuelve una cadena representativa de Persona"""
        return "%s: %s, %s %s, %s." % (
            self.__doc__[25:34], str(self.dni), self.nombre, 
            self.apellido, self.sexo)
```

```python
persona1 = MiembroClub("13458796V", "Leonardo", "Caballero", "M")
print(persona1)
```

```python
from abc import ABC, ABCMeta, abstractmethod
class Jugador(MiembroClub):
    """Clase que representa a un miembro del club"""

    def __init__(self, dni, nombre, apellido, sexo, club):
       

        # Invoca al constructor de clase MiembroCLub
        MiembroClub.__init__(self, dni, nombre, apellido, sexo)

        # Nuevos atributos
        self.club = club
       
    @abstractmethod
    def __str__(self):
        return "%s %s, club: '%s'." % (
            self.nombre, self.apellido, 
            self.club)
    @abstractmethod   
    def __eq__(self, other):
        print("___eq__ called: %r == %r ?" % (self, other))
        return self.value == other
```

```python
jugador1 = Jugador("12222696V", "Antonio", "Sanchez", "M","Vallecas")
print(jugador1)
```

```python
class Defensa(Jugador):
    """Clase que representa a un miembro del club"""

    def __init__(self, dni, nombre, apellido, sexo, rol):
       

        # Invoca al constructor de clase MiembroCLub
        MiembroClub.__init__(self, dni, nombre, apellido, sexo)

        # Nuevos atributos
        self.rol = rol
        self.tareas = ['Atajar balones ','Posesión','Salida de balón','Mantener Marcajes']

    def __str__(self):
        return "%s %s, rol: '%s', sus tareas: %s." % (
            self.nombre, self.apellido, 
            self.rol, self.consulta_tareas())

    def consulta_tareas(self):
        """Mostrar las tareas del Supervisor"""
        return ', '.join(self.tareas)
    
    def __eq__(self, other):
        print("Defensa __eq__ called: %r == %r ?" % (self, other))
        return self.value == other
```

```python
jugador1 = Defensa("12222696V", "Antonio", "Sanchez", "M","Defensa")
print(jugador1)
```

```python
from abc import ABC, ABCMeta, abstractmethod
class Delantero(Jugador):
    """Clase que representa a un miembro del club"""

    def __init__(self, dni, nombre, apellido, sexo, rol):
       

        # Invoca al constructor de clase MiembroCLub
        MiembroClub.__init__(self, dni, nombre, apellido, sexo)

        # Nuevos atributos
        self.rol = rol
        self.tareas = ['Desmarcarse ','Pases en profundidad','Incursiones al area','Disparos certeros']

        @abstractmethod
        def getNom(self):
            return self.__nom
    
        @abstractmethod    
        def setNom(self, nom):
            self.__nom=nom
    
    def __str__(self):
        return "%s %s, rol: '%s', sus tareas: %s." % (
            self.nombre, self.apellido, 
            self.rol, self.consulta_tareas())


    def consulta_tareas(self):
        """Mostrar las tareas del Supervisor"""
        return ', '.join(self.tareas)
    
    def __eq__(self, other):
        print("Delantero  __eq__ called: %r == %r ?" % (self, other))
        return self.value == other
```

```python
jugador2 = Delantero("1324211F", "Ramón", "Rodriguez", "M","Delantero Centro")
print(jugador2)
```
